```python
df2_2=df2_1.rename(columns={'Código CUT Comuna':'CODIGO COMUNA'})
df1_3=df1_2.rename(columns={'Codigo comuna':'CODIGO COMUNA'})
```

```python
dff=df1_3.merge(df2_2, on='CODIGO COMUNA')
```

```python
dff.head()
```

```python
dff['Fecha de Inicio'] = dff['Fecha de Inicio'].apply(lambda x: x.split(' ')[0])
dff['Fecha de Término'] = dff['Fecha de Término'].apply(lambda x: x.split(' ')[0])

dff['INICIO'] = pd.to_datetime(dff['Fecha de Inicio'], format='%Y/%m/%d')
dff['TERMINO'] = pd.to_datetime(dff['Fecha de Término'], format='%Y/%m/%d')
```

```python
dff.head()
```

```python
dff.drop('n_REGION',axis=1,inplace=True)
```

```python
dff.drop('Region_x',axis=1,inplace=True)
dff.drop('Comuna',axis=1,inplace=True)
dff.drop('Fecha de Inicio',axis=1,inplace=True)
dff.drop('Fecha de Término',axis=1,inplace=True)
dff.drop('Region_y',axis=1,inplace=True)
```
